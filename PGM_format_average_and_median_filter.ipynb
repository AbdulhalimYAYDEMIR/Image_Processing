```python
import matplotlib.pyplot as plt # görüntüyü bastırmak için gereklidir
import numpy as np # array işlemleri için gerekli
import math # matematiksel işlemler için gerekli



#Fig0335(a)(ckt_board_saltpep_prob_pt05).pgm dosyası 'rb' modunda açılarak okundu.
with open("Fig0335(a)(ckt_board_saltpep_prob_pt05).pgm", "rb") as fileObject:
    fileContent=fileObject.read()
    

    #dosyanın ilk iki karakterine bakılarak P5 formatı olup olmadığına bakıldı
    print('Dosya bilgisi :\n')
    i=0
    while fileContent[i]!=10:
        if i==0:
            if (fileContent[0]==80) & (fileContent[1]==53):
                print('P5 pgm formatı')  
            else:
                print('P5 pgm formatı değil')
        i=i+1

    
    
    
    
    # '#' ile başlayan satır geçildi.
    i=i+1
    while fileContent[i]!=10:
        if fileContent[i]==35:
            while fileContent[i]!=10:
                i=i+1
    
    
    
    
    
    # genişlik yükseklik bilgileri yazdırıldı
    i=i+1
    while fileContent[i]!=10:
        if fileContent[i]==35:
            while fileContent[i]!=10:
                i=i+1  
        else:
            string=[]
            
            while fileContent[i]!=32:
                width=chr(fileContent[i])
                string.append(width)       
                i=i+1
                
            string=''.join(string)
            width=int(string)
            print("width=",width)
            i=i+1
            string=[]
            
            while fileContent[i]!=10:
                
                height=chr(fileContent[i])
                string.append(height)
                i=i+1
                
            string=''.join(string)
            height=int(string)
            print("height=",height)
                   
             
               
    # maxval değeri yazdırıldı
    i=i+1
    string=[]
    while fileContent[i]!=10:
        maxval=chr(fileContent[i])
        string.append(maxval)       
        i=i+1
    
    string=''.join(string)
    maxval=int(string)
    print("maxval=",maxval)
    
    
    # piksel değerleri arraya geçirildi
    i=i+1
    b=i
    j=1
    f=np.arange(height*width).reshape((height,width))      
    
    for m in range(height):
        for n in range(width):
            f[m][n]=fileContent[i]
            j=j+1
            i=i+1
            
    

    # 3x3 matrislik işlem için gerekli veriler oluşturuldu
    interval=[3]   
    dosya_isim=["3.35(medyan. mask.-median. mask.3x3).pgm"]
    g=np.arange(height*width).reshape((height,width))     
    figure=["Figure17 - 3.35 medyan maskeleme 3*3"]
    counter=0
    # her pikselin etrafındaki 3x3 matrisi için değerler hesaplandı
    for d in interval:
        for m in range(height):
            for n in range(width):
                g[m][n]=0
                median=np.zeros(d*d)
                c=0
                for r in range(math.floor(-(d-1)/2),math.floor(((d-1)/2)+1)):
                    for s in range(math.floor(-(d-1)/2),math.floor(((d-1)/2)+1)):
                        if ((m+r)>(height-1)) | ((n+s)>(width-1)) | ((m+r)<0) | ((n+s)<0):
                            median[c]=0
                            c=c+1
                        else:
                            median[c]=f[m+r][n+s]
                            c=c+1
                            
                # medyan değerini bulmak için sıralama yapıldı 
                for k in range(d*d):
                    for l in range(d*d):
                        if l==8:
                            break
                        if (median[l+1]<median[l]):
                            temporary=median[l+1]
                            median[l+1]=median[l]
                            median[l]=temporary
                      
                # ortanca (medyan) değeri arraya aktarıldı
                g[m][n]=median[math.floor(((d*d)-1)/2)]
                    
    
        # sonuç matrisine göre dosya içeriği güncellendi  
        j=1
        i=b
        fileContent=bytearray(fileContent)
        for m in range(height):
            for n in range(width):
                fileContent[i]=g[m][n]
                j=j+1
                i=i+1
            
            
        # dosyaya yazdırılmak için byte tipine dönüştürüldü
        fileContent=bytes(fileContent)
    
    
    
        # dosyanın yeni hali 3.33(ortalama maskeleme-average mask3,5,9,15,35).pgm isimli dosyaya yazdırıldı
        with open(dosya_isim[counter], "wb") as fileObject:
            fileObject.write(fileContent)  
        
        
        # görüntü ekrana yazdırıldı
        plt.figure()
        plt.title(figure[counter])
        plt.imshow(g,cmap='gray',vmin=0, vmax=255)
        
        
        counter=counter+1
        
```
